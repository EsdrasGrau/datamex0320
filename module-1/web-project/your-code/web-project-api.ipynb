```python
#To clean the data 
def clean_data(df):
    df = df.copy()
    df = df.fillna('Unknown')
    df = df.drop(['countryCode', 'fixed', 'global', 'counties', 'launchYear', 'type'], axis = 1) 
    df = df.rename(columns = {"localName": "holidayjp", "name": "holiday" })
    df["date"]= pd.to_datetime(df["date"])
    return df
```
